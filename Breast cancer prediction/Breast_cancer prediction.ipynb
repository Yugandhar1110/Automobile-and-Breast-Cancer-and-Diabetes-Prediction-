```python
df = df.drop('radius_mean',axis = 1)
```

```python
input_col = df.drop('diagnosis',axis=1)
out_col = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_train,input_test,out_train,out_test = train_test_split(input_col,out_col,train_size=0.8,random_state=48)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import BernoulliNB,GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import NuSVC,SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
Model = [LogisticRegression,RandomForestClassifier,BernoulliNB,GaussianNB,KNeighborsClassifier,NuSVC,DecisionTreeClassifier]
```

```python
def model_train_tes_acc(Model):
    model = Model()
    model.fit(input_train,out_train)
    train_pred = model.predict(input_train)
    test_pred = model.predict(input_test)
    train_acc = accuracy_score(train_pred,out_train)
    test_acc = accuracy_score(test_pred,out_test)
    return (train_acc,test_acc)
```

```python
for i in Model:
    print(f'{str(i.__name__)}---> accuracy is {model_train_tes_acc(i)}')
```
